```python
import pandas as pd
import requests
from transformers import pipelines, AutoModelForTableQuestionAnswering, AutoTokenizer, TableQuestionAnsweringPipeline
```

```python
url = "https://mgmits.ac.in/b-tech-admissions-2021/" 
header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

req = requests.get(url, headers= header)

admission = pd.read_html(req.text, header=1)
```

```python
admission_data = admission[0]
```

```python
admission_data
```

```python


# admission_data[admission_data.notna()]
admission_data.columns
admission_data = admission_data.astype("str")
admission_data.info()
```

```python
tokenizer = AutoTokenizer.from_pretrained("google/tapas-medium-finetuned-wtq")

model = AutoModelForTableQuestionAnswering.from_pretrained("google/tapas-medium-finetuned-wtq")
```

```python
nlp = TableQuestionAnsweringPipeline(model = model, tokenizer=tokenizer)
```

```python
result = nlp(table = admission_data, query = ["What are the different branches"])
# result = nlp(table = admission_data, query = ["Number of quota for muthoot employees quota in computer science and engineering"])
# result = nlp(table = admission_data, query = ["Total in computer science and engineering"])
```
