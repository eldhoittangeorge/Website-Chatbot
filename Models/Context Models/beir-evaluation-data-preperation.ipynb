```python
# Create queries
queries_file = []

def create_questoin(data):
    _id = temp_dict[data._id['$oid']]
    print(_id)
    questions = data.questions
    tmp_list = []
    for index, question in enumerate(questions):
        tmp_list.append({"_id":str(_id)+"~"+str(index), "text": question})

    return tmp_list

queries_file = data.apply(create_questoin, axis=1)
```

```python
final_queries_file = sum(queries_file, [])
```

```python
queries_df = pd.DataFrame(final_queries_file) 
# queries_df.to_json("Data/queries.json")
queries_df._id = queries_df.index
```

```python
import csv

qrels = {}

reader = csv.reader(open("Data/qrels.tsv", encoding="utf-8"), delimiter="\t", quoting=csv.QUOTE_MINIMAL)
next(reader)
for id_, row in enumerate(reader):
    query_id, corpus_id, score = row[0], row[1], int(row[2])

    if query_id not in qrels:
        qrels[query_id] = {corpus_id: score}
    else:
        qrels[query_id][corpus_id] = score
```
